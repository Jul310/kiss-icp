```python
import folium
from tools.tum_tools import load_gt_poses
from tools.tum_definitions import *
import os
import numpy as np
```

```python
ds = "mdv3_1"
data_path = f"/home/julian/projects/thesis/datasets/gps_data/{ds}"
```

```python
os.path.exists(data_path)
```

```python
poses, _ = load_gt_poses(data_path)
```

```python
N = len(poses)
M = 13
segement_size = 3*N // (4*(M-1) - 4)
print(segement_size)

segement_size = int(N // (M -1 + 0.3333333))
print(segement_size)

segments = []

cnt = 0
for i in range(M-1):
    segments.append(poses[cnt:cnt+segement_size])
    cnt += segement_size
else:
    segments.append(poses[cnt:])
    
```

```python
151 / 500
```

```python
m = folium.Map(location=[poses[0,0], poses[0,1]], zoom_start=16, )

gray = "#333333"
blue = "#0065BD"
orange = "#E37222"

# folium.PolyLine(poses[1:,:2], color="blue", weight=2.5, opacity=1).add_to(m)
for i, pose in enumerate(segments):
    if i == 4:
        folium.PolyLine(pose[1:,:2], color=orange, weight=5, opacity=1).add_to(m)
    elif i == 10:
        folium.PolyLine(pose[1:,:2], color=blue, weight=5, opacity=1).add_to(m)
    else:
        folium.PolyLine(pose[1:,:2], color=gray, weight=2.5, opacity=1).add_to(m)

m.fit_bounds([np.min(poses), np.max(poses)])
```
